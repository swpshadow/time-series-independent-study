```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./hungary_chickenpox/hungary_chickenpox.csv")
```

```python
df
```

```python
plt.plot(df["BACS"])
```

```python
def moving_average(data, window = 3):
    res = []
    for i in range(window, len(data) - window):
        res.append(sum(data[i-window:i])/window)
    return res
```

```python

res = moving_average(df["BACS"], window = 3)
plt.plot(res)
```

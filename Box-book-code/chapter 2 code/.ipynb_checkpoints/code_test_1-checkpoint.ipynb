```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./hungary_chickenpox/hungary_chickenpox.csv")
```

```python
plt.plot(df["BACS"])
```

```python
def moving_average(data, window = 3, weighted = False):
    res = []
    if weighted:
        weights = [ i for i in range(window) ]
    else:
        weights = [1 for i in range(window) ]
    weights = [x / sum(weights) for x in weights]
    for i in range(window, len(data) - window):
        res.append(np.dot(data[i-window:i], weights)/window)
    return res
```

```python

res = moving_average(df["BACS"], window = 3)
plt.plot(res)
```
